```python
#second project

url='https://www.forbesindia.com/article/explainers/top-10-richest-people-india/85909/1'
page=requests.get(url)
data=BeautifulSoup(page.text,'html')
print(data)
```

```python
title=data.find('title')
print(title.text)
```

```python
table=data.find_all('table')
print(table)
```

```python
heading=data.find_all('th')
print(heading)
```

```python

head=[i.text for i in heading] # list comphrehension : list name=[output for loop]
#for i in heading:
print(head)
```

```python
#create a data frames
import pandas as pd
df=pd.DataFrame(columns=head)

print(df)
```

```python
ind = 0
rows = data.find_all('tr')
#print(rows)
for i in rows[1:9]:
  #print(i)
  erd = i.find_all('td')
  er = [j.text for j in erd]
  #print(er)
  df.loc[ind] = er
  ind += 1
df
```

```python
print(df)
df.to_csv("top 10 richest mans in india.csv")
```
