```python
import pandas as pd
from sklearn.linear_model import LinearRegression
data = pd.DataFrame({
    "Age": [25, 30, 35, 40, 45],
    "Height": [170, 175, 180, 185, 190],
    "Weight": [65, 70, 75, 80, 85],
    "Blood Pressure": [120, 125, 130, 135, 140]
})
X = data[["Age", "Height"]]
y = data["Blood Pressure"]
model = LinearRegression()
model.fit(X, y)
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
new_data = pd.DataFrame({"Age": [32], "Height": [182]})
predictions = model.predict(new_data)
print("Predicted Blood Pressure:", predictions[0])
```

```python
from fpdf import FPDF
import pandas as pd

# Convert TXT to CSV
def txt_to_csv(input_file, output_file):
    # Read the text file into a DataFrame
    with open(input_file, 'r', encoding='utf-8') as file:
        content = file.readlines()
    df = pd.DataFrame(content, columns=['content'])

    # Write DataFrame to CSV
    df.to_csv(output_file, index=False)

# Convert CSV to PDF
def csv_to_pdf(input_file, output_file):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(input_file)

    # Create a PDF object
    pdf = FPDF()
    pdf.add_page()

    # Set font for the PDF
    pdf.set_font("Arial", size=12)

    # Add content from DataFrame to PDF
    for index, row in df.iterrows():
        # Convert Unicode characters to UTF-8 to avoid encoding errors
        content = row['content'].encode('latin-1', 'replace').decode('latin-1')
        pdf.cell(200, 10, txt=content, ln=True)

    # Save the PDF file
    pdf.output(output_file)

# Example usage
txt_to_csv("/content/whatsapp.txt", "/content/whatsapp.csv")
csv_to_pdf("/content/whatsapp.csv", "/content/whatsapp.pdf")
```
